```python
import pandas as pd
from plotnine import *
from plotnine.data import mpg
from plotnine.data import mtcars
```

```python
#grafico 1

(ggplot(mpg, aes(x = "class", fill = "trans")) +
 geom_bar(position = "fill")+
 labs(x = "Tipo de carro", 
      y = "Quantidade" )+
 coord_flip()+
 theme_classic()
 )
```

```python
#grafico 2
(ggplot(mtcars, aes(x="hp",
                    y="qsec", size= "carb", fill = "cyl")) +
 geom_point()+
 labs(x = "Potência em Cavalos",
      y = "Consumo de combusítvel"
   
)+
 theme_classic ()
)


#Podemos perceber que quanto maior a potência, menos o consumo de combustível. Quanto maior o número de cilindradas, maior a potência.
```
